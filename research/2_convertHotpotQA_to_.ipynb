## ‚úÖ Code Python chuy·ªÉn ƒë·ªïi:

```python
def convert_hotpotqa_example(example):
    # L·∫•y danh s√°ch ti√™u ƒë·ªÅ c·∫ßn h·ªó tr·ª£
    supporting_titles = {title for title, _ in example["supporting_facts"]}
    
    # T·∫°o danh s√°ch context m·ªõi
    new_contexts = []
    for idx, (title, sentences) in enumerate(example["context"]):
        paragraph_text = " ".join(sentences)
        is_supporting = title in supporting_titles
        new_contexts.append({
            "idx": idx,
            "title": title,
            "paragraph_text": paragraph_text,
            "is_supporting": is_supporting
        })
    
    # Tr·∫£ v·ªÅ dict m·ªõi theo format ƒë√£ chuy·ªÉn
    return {
        "question_id": example["_id"],
        "question": example["question"],
        "contexts": new_contexts,
        "answer": example["answer"]
    }
```
